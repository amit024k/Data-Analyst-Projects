## Questions for analysis

1. Do a greater number of restaurants provide online delivery as opposed to offline services?
2. Which types of restaurants are the most favored by the general public?
3. What price range is preferred by couples for their dinner at restaurants?

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Zomato data .csv')
```

```python
df.head(10)
```

```python
df.info()
```

- There is no null value in data

```python
modeTotal = df['online_order'].value_counts()
```

```python
plt.figure()
plt.bar(modeTotal.index, modeTotal.values)
plt.xlabel('Online order or not')
plt.ylabel('Total counts')
plt.title('Online vs offline')
plt.show()
```

## Answer 1
No, the restaurants which provide online delivery are lesser as compared to offline services

```python
totalCount = df.groupby('listed_in(type)')['votes'].sum()
```

```python
plt.figure()
plt.bar(totalCount.index, totalCount.values)
plt.xlabel('Category')
plt.ylabel('Total count of votes')
plt.title('Categorywise votes')
plt.show()
```

## Answer 2 
Dining type of restuarant is more favoured in general public as shown in above diagram

```python
costCountTotal = df['approx_cost(for two people)'].value_counts()
```

```python
plt.figure()
plt.bar(costCountTotal.index, costCountTotal.values)
plt.xlabel('Price')
plt.ylabel('Number of customers')
plt.title('Customers vs Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='approx_cost(for two people)', data=df)

# Set title and labels
plt.title('Price and number of customers')
plt.xlabel('price')
plt.ylabel('Count')

# Show the plot
plt.show()
```

## Answer 3
Restaurants having price of Rs.300 are preferred among customers
